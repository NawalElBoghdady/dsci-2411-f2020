# Installation Instructions for running R on Jupyter Lab (Windows Installation)

For this course, we will be using both R and Python. These are the installation instructions for getting the R kernel up and running on Jupyter Notebooks

For a detailed set of instructions, visit: https://richpauloo.github.io/2018-05-16-Installing-the-R-kernel-in-Jupyter-Lab/

1. Open Anaconda Prompt
2. Locate where you have the R kernel installed. This is NOT the directory where you have R studio. On my machine, it is: <code>C:\Program Files\R\R-3.5.2\bin</code>
3. On the Anaconda Prompt, cd to the location of where your R kernel is. Here is the command I enter from my own computer:  <code>cd C:\Program Files\R\R-3.5.2\bin</code>
4. Type this command <code>install.packages("devtools")</code> to get you the devtools which will allow you to use the <code>install_github()</code> command we will use in the next steps. Note that the <code>install.packages("devtools")</code> is an R command.
5. Before fetching the IRkernel from github, you will need to make sure you have git already installed on your machine, otherwise you will get an error. You can download git from here: https://git-scm.com/download/win
6. Run and install git. I used all the default settings
7. In the Anaconda Prompt running R, type this command: <code>devtools::install_github("IRkernel/IRkernel")</code>. Because I was working with R previously, I had installed a number of packages. I received a prompt that asked me to update some of them. I decided to update all of them. If you get an error here that there is a problem removing an older installation of a certain package, go to the following directory on your computer <code>C:\Users\nawal_igk9s0a\Documents\R\win-library\3.5</code> (wherever your R packages are installed) and **manually** delete these problematic packages, along with devtools. Restart R on Anaconda, and then re-install these problematic packages using the <code>install.packages("[package_name]")</code> command, followed by <code>install.packages("devtools")</code>. Next, run the command <code>devtools::install_github("IRkernel/IRkernel")</code>. This worked for me. 
8. Finally, run the command <code>IRkernel::installspec()</code>. This should work without error.
9. Quit R, and run the command <code>jupyter lab</code> to find the newly installed R Kernel.
